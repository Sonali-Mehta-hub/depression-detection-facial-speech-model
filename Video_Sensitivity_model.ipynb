```python
import os
import cv2
import numpy as np

dataset_path = '/content/drive/MyDrive/fer2013_dataset/train'

depressed = ['sad', 'fear', 'angry']
normal = ['happy', 'neutral', 'surprise']

X = []  # images
y = []  # labels

IMG_SIZE = 48  # FER2013 images 48x48 hoti hain

# Depressed images load karo
for emotion in depressed:
    folder = os.path.join(dataset_path, emotion)
    for img_name in os.listdir(folder):
        img_path = os.path.join(folder, img_name)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if img is not None:
            img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
            X.append(img)
            y.append(0)  # 0 = Depressed

# Normal images load karo
for emotion in normal:
    folder = os.path.join(dataset_path, emotion)
    for img_name in os.listdir(folder):
        img_path = os.path.join(folder, img_name)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        if img is not None:
            img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
            X.append(img)
            y.append(1)  # 1 = Normal

print(f"Total images: {len(X)}")
print(f"Total labels: {len(y)}")
```
